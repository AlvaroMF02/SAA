```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# generasmo valores de horas de sueo de 0 a 8
horas_sueo = np.linspace(0, 8, 50)

# definimos la relacin lineal
pendiente = 9.8
intercepto = 20
y = pendiente * horas_sueo + intercepto

# Aadir ruido con diferentes niveles de dispersin
np.random.seed(42)
desviacion_baja = 5    # Baja dispersin
desviacion_alta = 15   # Alta dispersin

energia_baja_ruido = y + np.random.normal(0, desviacion_baja, size=horas_sueo.shape)
energia_alta_ruido = y + np.random.normal(0, desviacion_alta, size=horas_sueo.shape)

# Creamos un DataFrame para almacenar los datos
# a energa sigue la tendencia lineal =12+20
datos = pd.DataFrame({
    "Horas_de_sueo": horas_sueo,
    "Energa_baja_dispersin": energia_baja_ruido,
    "Energa_alta_dispersin": energia_alta_ruido
})

"""
Primera fila (0 horas de sueo):

Horas_de_sueo: 0 horas.
Energa_baja_dispersin: 22.48, lo que significa que con 0 horas de sueo, la persona percibe un 22.48% de energa debido al ruido de baja dispersin.
Energa_alta_dispersin: 24.86, lo que refleja el mismo caso pero con mayor variabilidad en la energa percibida.

Segunda fila (0.163 horas de sueo, aprox. 10 minutos):
Horas_de_sueo: Aproximadamente 0.16 horas.
Energa_baja_dispersin: 21.27, que refleja la energa percibida despus de aproximadamente 10 minutos de sueo, con baja dispersin.
Energa_alta_dispersin: 16.18, con ms variabilidad en el valor de la energa debido al ruido alto.
"""

# vemos los datos generados
plt.figure(figsize=(10, 6))
plt.plot(horas_sueo, y, label="Relacin lineal verdadera (sin ruido)", color="black", linestyle="-")
plt.scatter(horas_sueo, energia_baja_ruido, label="Con baja dispersin ( = 5)", color="magenta", alpha=0.7)
plt.scatter(horas_sueo, energia_alta_ruido, label="Con alta dispersin ( = 15)", color="blue", alpha=0.7)
plt.xlabel("Horas de sueo (h)")
plt.ylabel("Energa percibida (%)")
plt.title("Relacin entre horas de sueo y energa percibida")
plt.legend()
plt.grid()
plt.show()

# Mostrar una muestra de los datos generados
print(datos.head(50))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

horas_sueo = np.linspace(0, 8, 50)

pendiente = 12
intercepto = 20
y = pendiente * horas_sueo + intercepto





# Definir la desviacin estndar del ruido que aumenta con x
# A medida que x crece, la desviacin estndar del ruido tambin lo hace
np.random.seed(42)
desviacion_base = 5
desviacion_ruido = desviacion_base * horas_sueo

# Aadir ruido con la desviacin estndar dependiente de x
energia_con_ruido = y + np.random.normal(0, desviacion_ruido)

# Crear un DataFrame para almacenar los datos
datos = pd.DataFrame({
    "Horas_de_sueo": horas_sueo,
    "Energa_con_ruido": energia_con_ruido
})

# Visualizar los datos generados
plt.figure(figsize=(10, 6))
plt.plot(horas_sueo, y, label="Relacin lineal verdadera (sin ruido)", color="black", linestyle="-")
plt.scatter(horas_sueo, energia_con_ruido, label="Con dispersin creciente ( =  * x)", color="orange", alpha=0.7)
plt.xlabel("Horas de sueo (h)")
plt.ylabel("Energa percibida (%)")
plt.title("Relacin entre horas de sueo y energa percibida con dispersin creciente")
plt.legend()
plt.grid()
plt.show()

# Mostrar una muestra de los primeros 10 datos generados
print(datos.head(50))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

horas_sueo = np.linspace(0, 8, 50)

pendiente = 12
intercepto = 20
y = pendiente * horas_sueo + intercepto




# (homocedasticidad)
np.random.seed(42)
desviacion_constante = 5  # Ruido constante
energia_con_ruido_constante = y + np.random.normal(0, desviacion_constante, size=horas_sueo.shape)

# (heterocedasticidad)
desviacion_ruido = desviacion_constante * horas_sueo  # Aumento del ruido con x
energia_con_ruido_heterocedasticidad = y + np.random.normal(0, desviacion_ruido)

# Realizar la regresin lineal sobre los datos con dispersin constante (homocedasticidad)
regresion_constante = LinearRegression()
X = horas_sueo.reshape(-1, 1)
regresion_constante.fit(X, energia_con_ruido_constante)

# Realizar la regresin lineal sobre los datos con dispersin creciente (heterocedasticidad)
regresion_heterocedastica = LinearRegression()
regresion_heterocedastica.fit(X, energia_con_ruido_heterocedasticidad)

# Visualizar los resultados
plt.figure(figsize=(10, 6))

# Datos con dispersin constante (homocedasticidad)
plt.scatter(horas_sueo, energia_con_ruido_constante, label="Datos con dispersin constante", color="blue", alpha=0.7)
plt.plot(horas_sueo, regresion_constante.predict(X), label="Regresin lineal (dispersin constante)", color="green", linestyle="-")

# Datos con dispersin creciente (heterocedasticidad)
plt.scatter(horas_sueo, energia_con_ruido_heterocedasticidad, label="Datos con dispersin creciente", color="orange", alpha=0.7)
plt.plot(horas_sueo, regresion_heterocedastica.predict(X), label="Regresin lineal (heterocedasticidad)", color="red", linestyle="-")

plt.xlabel("Horas de sueo (h)")
plt.ylabel("Energa percibida (%)")
plt.title("Regresin lineal sobre datos con dispersin constante y heterocedasticidad")
plt.legend()
plt.grid()
plt.show()

# Mostrar los coeficientes de las regresiones
print("Regresin sobre dispersin constante (homocedasticidad):")
print(f"Pendiente: {regresion_constante.coef_[0]:.2f}")
print(f"Intercepto: {regresion_constante.intercept_:.2f}")
print(f"Error cuadrtico medio (MSE): {mean_squared_error(energia_con_ruido_constante, regresion_constante.predict(X)):.2f}\n")

print("Regresin sobre dispersin creciente (heterocedasticidad):")
print(f"Pendiente: {regresion_heterocedastica.coef_[0]:.2f}")
print(f"Intercepto: {regresion_heterocedastica.intercept_:.2f}")
print(f"Error cuadrtico medio (MSE): {mean_squared_error(energia_con_ruido_heterocedasticidad, regresion_heterocedastica.predict(X)):.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

horas_sueo = np.linspace(0, 8, 50)

pendiente = 12
intercepto = 20
c = 5
# y = pendiente * horas_sueo + intercepto     # normal
y = pendiente * horas_sueo + intercepto + c * horas_sueo * horas_sueo    # para que se cree una parabola

# (homocedasticidad)
np.random.seed(42)
desviacion_constante = 5  # Ruido constante
energia_con_ruido_constante = y + np.random.normal(0, desviacion_constante, size=horas_sueo.shape)

# (heterocedasticidad)
desviacion_ruido = desviacion_constante * horas_sueo  # Aumento del ruido con x
energia_con_ruido_heterocedasticidad = y + np.random.normal(0, desviacion_ruido)

# regresin lineal sobre los datos con dispersin constante (homocedasticidad)
regresion_constante = LinearRegression()
X = horas_sueo.reshape(-1, 1)  # Convertir a formato 2D para sklearn
regresion_constante.fit(X, energia_con_ruido_constante)

# regresin lineal sobre los datos con dispersin creciente (heterocedasticidad)
regresion_heterocedastica = LinearRegression()
regresion_heterocedastica.fit(X, energia_con_ruido_heterocedasticidad)





# Calcular los valores predichos para ambos casos
predicciones_constante = regresion_constante.predict(X)
predicciones_heterocedastica = regresion_heterocedastica.predict(X)

#Calcular los residuales para ambos casos
residuales_constante = energia_con_ruido_constante - predicciones_constante
residuales_heterocedastica = energia_con_ruido_heterocedasticidad - predicciones_heterocedastica

# Graficar los residuales de ambos modelos
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.scatter(horas_sueo, residuales_constante, color="magenta", label="Residuales (dispersin constante)")
plt.axhline(0, color="black", linestyle="--")
plt.xlabel("Horas de sueo (h)")
plt.ylabel("Residuales")
plt.title("Residuales - Dispersin constante")
plt.grid()

plt.subplot(1, 2, 2)
plt.scatter(horas_sueo, residuales_heterocedastica, color="blue", label="Residuales (heterocedasticidad)")
plt.axhline(0, color="black", linestyle="--")
plt.xlabel("Horas de sueo (h)")
plt.ylabel("Residuales")
plt.title("Residuales - Heterocedasticidad")
plt.grid()

plt.tight_layout()
plt.show()

# Crear histogramas de los residuales
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.hist(residuales_constante, bins=15, color="magenta", alpha=0.7)
plt.title("Histograma de residuales - Dispersin constante")
plt.xlabel("Residuales")
plt.ylabel("Frecuencia")

plt.subplot(1, 2, 2)
plt.hist(residuales_heterocedastica, bins=15, color="blue", alpha=0.7)
plt.title("Histograma de residuales - Heterocedasticidad")
plt.xlabel("Residuales")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()



# Graficar la recta original y las rectas ajustadas
plt.figure(figsize=(10, 6))

# Graficar la relacin lineal original
plt.plot(horas_sueo, y, label="Relacin lineal verdadera", color="black", linestyle="-")

# Graficar la recta ajustada con dispersin constante
plt.plot(horas_sueo, predicciones_constante, label="Recta ajustada (dispersin constante)", color="magenta", linestyle="-")

# Graficar la recta ajustada con heterocedasticidad
plt.plot(horas_sueo, predicciones_heterocedastica, label="Recta ajustada (heterocedasticidad)", color="blue", linestyle="-")

plt.scatter(horas_sueo, energia_con_ruido_constante, color="magenta", alpha=0.5, label="Datos con dispersin constante")
plt.scatter(horas_sueo, energia_con_ruido_heterocedasticidad, color="blue", alpha=0.5, label="Datos con heterocedasticidad")

plt.xlabel("Horas de sueo (h)")
plt.ylabel("Energa percibida (%)")
plt.title("Comparacin de la recta original y ajustadas")
plt.legend()
plt.grid()
plt.show()
```
